```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Mahesh/Downloads/industry.csv")
print(df)
```

```python
data=df['Industry']
val=df['Value']
plt.figure(figsize=(50, 50))
plt.bar(data,val,label='industries',color='g')
plt.legend()
plt.xlabel('range')
plt.ylabel('names')
plt.title('bar plot')
plt.xticks(rotation=45) 
plt.show()
```

```python
plt.figure(figsize=(30, 30))
plt.pie(val, labels=data,startangle=90,shadow=True, autopct='%1.1f%%')
plt.title('pie chart')
plt.show()
```

```python
plt.figure(figsize=(50, 25))
sns.boxplot(data=df,x=data, y=val)
plt.xlabel('industry')
plt.ylabel('value')
plt.title('Boxplot ')
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(50, 25))
plt.scatter(data,val)
plt.xlabel('industry')
plt.ylabel('value')
plt.title('scatter')
plt.xticks(rotation=45)  
plt.show()
```
