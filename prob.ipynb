```python
import numpy as np
import matplotlib.pyplot as plt
import random
import seaborn as sns
import pandas as pd
def pick(n):
    result=[]
    for i in range(1,n+1):
        result.append(random.choice(['love candy','love soda','dont love candy','dont love soda']))
    return result
result=pick(14)
print(result)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=result)
plt.title('Graph')
plt.xlabel('items')
plt.ylabel('range')
plt.show()
```

```python
phrase_counts = pd.Series(result).value_counts()

plt.figure(figsize=(6, 6))
plt.pie(phrase_counts, labels=phrase_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribution of Phrases")
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=phrase_counts.index, y=phrase_counts.values, palette="Blues")
plt.xlabel("Phrases")
plt.ylabel("Count")
plt.title("Bar Plot of Phrase Distribution")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(phrase_counts.index, phrase_counts.values, marker='o', linestyle='-')
plt.xlabel("Phrases")
plt.ylabel("Count")
plt.title("Bar Plot of Phrase Distribution")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(phrase_counts.index, phrase_counts.values)
plt.xlabel("Phrases")
plt.ylabel("Count")
plt.title("Bar Plot of Phrase Distribution")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
colors = ['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon']
plt.stackplot(phrase_counts.index, [phrase_counts.get('love candy', 0),
                                    phrase_counts.get('love soda', 0),
                                    phrase_counts.get('dont love candy', 0),
                                    phrase_counts.get('dont love soda', 0)],
                                    labels=['love candy', 'love soda', 'dont love candy', 'dont love soda'],
                                    colors=colors)
plt.xlabel("Phrases")
plt.ylabel("Count")
plt.title("Stack Plot of Phrase Distribution")
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
