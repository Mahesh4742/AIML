```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5]
sleep=[8,6,4,7,5,9]
work=[2,5,6,3,4,2]
eat=[7,8,6,4,5,2]
play=[8,5,7,4,12]
plt.plot([],[],color='m', label='sleep', linewidth=5)
plt.plot([],[],color='m', label='working', linewidth=5)
plt.plot([],[],color='m', label='eating', linewidth=5)
plt.plot([],[],color='m', label='playing', linewidth=5)
plt.stackplot(days, sleep, work, eat, play, colors=['m','c','r','k'])
plt.plt.xlabel('range')
plt.ylabel('length')
plt.title('Demo')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
slice=[7,2,2,13]
a=['sleep','eat','work','play']
c=['m','c','r','b']
plt.pie(slice, labels=a, colors=c, startangle=90,explode=(0,0.1,0,0),shadow=True, autopct='%1.1f%%')
plt.title('pie chart')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(t):
  return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.subplot(211)
plt.plot(t1, f(t1),'bo',t2,f(t2))

plt.subplot(212)
plt.plot(t2, np.cos(2+np.pi*t2))
plt.show()
```
