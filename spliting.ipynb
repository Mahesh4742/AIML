```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate synthetic data
np.random.seed(0)
X = np.random.rand(50,2)* 10

# Create a K-means model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)

# Fit the K-means model to the data
kmeans.fit(X)

# Get cluster labels and cluster centers
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

# Plot the data points and cluster centers
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='plasma')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='X', s=100, label='Cluster Centers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('K-means Clustering')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LinearRegression

# Generate synthetic data
np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = 2 * X + 1 + np.random.randn(100, 1)

# Initialize Leave-One-Out Cross-Validation
loo = LeaveOneOut()

# Initialize an array to store RMSE for each iteration
rmse_values =  []

# Perform Leave-One-Out Cross-Validation
for train_index, test_index in loo.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Fit a linear regression model on the training data
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Make predictions on the test data point
    prediction = model.predict(X_test)

    
    # Calculate RMSE for this iteration
    rmse = np.sqrt((prediction - y_test) ** 2)
    rmse_values.append(rmse)

# Flatten the RMSE values array
rmse_values = np.array(rmse_values).flatten()

# Plot the RMSE values for each iteration
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(rmse_values) + 1), rmse_values, marker='o', linestyle='-')
plt.xlabel('Iteration')
plt.ylabel('RMSE')
plt.title('Leave-One-Out Cross-Validation with Linear Regression')
plt.grid()
plt.show()

# Calculate the average RMSE
average_rmse = np.mean(rmse_values)
print(f'Average RMSE: {average_rmse:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Generate synthetic data
np.random.seed(0)
X = np.random.rand(50, 1) * 10
y = 2 * X + 1 + np.random.randn(50, 1)

# Split the data into a training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Fit a linear regression model on the training data
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)

# Create a scatter plot for the linear regression
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.scatter(X_test, predictions, color='red', marker='x', label='Predicted')
plt.plot(X_test, predictions, color='green', linewidth=2, label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression with Holdout Method')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate synthetic data
np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = 2 * X + 1 + np.random.randn(100, 1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = X[:80], X[80:], y[:80], y[80:]

# Fit a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)

# Calculate squared errors for each prediction
squared_errors = (predictions - y_test) ** 2

# Calculate the overall RMSE
rmse = np.sqrt(mean_squared_error(y_test, predictions))

# Create a scatter plot for the linear regression with RMSE
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.scatter(X_test, predictions, color='red', marker='x', label='Predicted')
plt.plot(X_test, predictions, color='green', linewidth=2, label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title(f'Linear Regression (RMSE: {rmse:.2f})')
plt.legend()
plt.grid()
plt.show()
```
